# Melt function   : 




```python
import numpy as np
import pandas as pd
data={
      "location":["Banglore","Mumbai","Delhi"],
      "Rainfall":["YES","NO","MAYBE"],
      "Jan":[25,24,19],
      "March":[14,16,14],
      "Feb":[4,27,14]
      }
x=pd.DataFrame(data)
print(x)
print()
x1=pd.melt(x,id_vars=["location","Rainfall"],var_name="Month",value_name="Date",value_vars=["Feb","Jan","March"])
print(x1)
print()
print()
x2=pd.melt(x,id_vars="location",var_name="Month",value_name="Date",value_vars=["Rainfall","Feb","Jan","March"])
print(x2)
print()
print()
x3=pd.melt(x,id_vars=["location"],var_name="Month",value_name="Date",value_vars=["Jan"])
print(x3)
```

# Groupby function


```python
import numpy as np
import pandas as pd
data={
      "Date":[5117,5117,5217,5217,5117,5217,5117,5217],
      "City":["NY","NY","NY","NY","ND","ND","ND","ND"],
      "Temp":[32,32,32,44,42,42,44,32],
      "Humidity":[54,54,54,67,56,56,74,72],
      
      }
x=pd.DataFrame(data)
print(x)
print()
print()
x1=x.groupby("City")
print(x1)
```

# Map fuction

```python
data={
      "Date":[5117,5117,5217,5217,5117,5217,5117,5217],
      "City":["NY","NY","NY","NY","ND","ND","ND","ND"],
      "Temp":[32,32,32,44,42,42,44,32],
      "Humidity":[54,54,54,67,56,56,74,72],
      
      }
x=pd.DataFrame(data)
print(x)
print()
print()
Season={"NY":"BRING BLANKETS","ND":"HOT LIKE HELL"}
x["Season"]=x["City"].map(Season)     #mapping season to cities
x
```
