### a) A table containing the columns â€˜Countryâ€™, â€˜Yearâ€™, â€˜Total suicidesâ€™. Total Suicides has to be calculated from the existing table.

```python
new_table=data.groupby(by=["country","year"]).sum()
mn=pd.DataFrame(new_table["suicides_no"])
ps=mn.rename(columns={"suicides_no":"Total Suicides"})
ps.head()
```

### b) A Table containing the columns â€˜Countryâ€™, â€˜Yearâ€™, â€˜per capita gdpâ€™.

```python
neww=pd.DataFrame(data[["country","year","gdp_per_capita ($)"]])
neww.head()
```

### c) Merge the above two tables using â€˜Countryâ€™ as the merge column.

```python
pd.merge(ps,neww, how ='inner', on ='country') 
```
