```python
#### Take n arguments and print even nos. from those

def evenfn(*args):
    x=[*args]          #### this statement is used to store any no. of element list in a variable
    for i in range(0,len(x)):
        if x[i]%2==0:
            print(x[i])
```

```python
evenfn(12,25,258,14)
```

```python
evenfn(87894,86413,1246,25,23,29,13)
```

```python
# Take two words and return true if begin with same alphabet
def wordcheck(a,b):    
    x=list(a)
    y=list(b)
    if x[0]==y[0]:
        return(True)
    else:
        return(False)
```

```python
wordcheck("Tarun","Tyagi")
```

```python
wordcheck("Nikhil","Tyagi")
```
