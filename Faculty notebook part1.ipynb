```python
Image('7.png')
```

```python
Data  = [3,5,2,8,9,2,7,2,2,2]
```

```python
np.mean(Data)
```

```python
np.median(Data)
```

```python
statistics.mode(Data)
```

```python
mean=np.mean(Data)
median=np.median(Data)
mode=statistics.mode(Data)
plt.figure(figsize=(10,5))
plt.hist(Data,bins=15,color='#46f2d0')
plt.axvline(mean,color='#ed0909',label='Mean')
plt.axvline(median,color='#2a22c9',label='Median')
plt.axvline(mode,color='green',label='Mode')
plt.xlabel('Data')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```
