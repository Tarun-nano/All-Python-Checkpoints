```python
import pandas as pd
import numpy as np
train=pd.read_csv("G:\Datasets\House_Price_Comp\housetrain.csv")
test=pd.read_csv("G:\Datasets\House_Price_Comp\housetest.csv")
pd.set_option("display.max_rows",None)
pd.set_option("display.max_columns",None)
```

```python
new_train=train.drop("Id",axis=1)
new_test=test.drop("Id",axis=1)
```

```python
dummytrain=pd.get_dummies(new_train)
dummytest=pd.get_dummies(new_test)
```

```python
dummytrain.shape
```

```python
dummytest.shape
```

```python
l1 = list(set(dummytrain.columns) - set(dummytest.columns))
dummytrain1 = dummytrain.drop(l1, axis = 1)
```

```python
dummytrain1["SalePrice"]=dummytrain["SalePrice"]
```

```python
dummytrain1.head()
```

```python
x=dummytrain1.drop("SalePrice",axis=1)
y=dummytrain1.SalePrice
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
scaled_train = pd.DataFrame(sc.fit_transform(dummytrain1), columns=dummytrain1.columns)
scaled_train.head()


```

```python
scaled_test = pd.DataFrame(sc.fit_transform(dummytest), columns=dummytest.columns)
scaled_test.head()
```

```python
x=scaled_train.drop("SalePrice",axis=1)
y=scaled_train.SalePrice
```

```python
from sklearn.linear_model import LinearRegression
Devil = LinearRegression()

```

```python
Devil.fit(pd.DataFrame(x),y)
```

```python
pred1=Devil.predict(scaled_test)
```

```python
pred1
```

```python
newHprice = sc.fit_transform(pd.DataFrame(train.SalePrice))
```

```python
hprice = sc.inverse_transform(pred1)
```

```python
hprice
```

```python
solution = pd.DataFrame({"Id":test.Id, "SalePrice":hprice})
```

```python
solution.to_csv("my_LR Model.csv", index = False)
```
