```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("G:\\Datasets\\UpVotes\\train_upvotes.csv")
test=pd.read_csv("G:\\Datasets\\UpVotes\\test_upvotes.csv")
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data.Tag.unique()
```

```python
z=pd.get_dummies(data["Tag"])
z.head()
```

```python
data1=pd.concat([data, z],axis=1)
```

```python
data=data1.drop(["Tag","ID"],axis=1)
```

```python
data.head()
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
scaled_train = pd.DataFrame(sc.fit_transform(data), columns=data.columns)
scaled_train.head()
```

```python
from sklearn.model_selection import train_test_split
y=scaled_train["Upvotes"]
x=scaled_train.drop(["Upvotes"],axis=1)
```

```python
from sklearn.linear_model import LinearRegression
Baby=LinearRegression()
```

```python
Baby.fit(x,y)
```

```python
test.head()
```

```python
scaled_test = pd.DataFrame(sc.fit_transform(test), columns=test.columns)
scaled_test.head()
```

```python
pred=Baby.predict(scaled_test)
```
