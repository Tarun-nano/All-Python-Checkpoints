<span style="font-family: Arial; font-weight:bold;font-size:1.9em;color:#03b7b4;"> 1.12  Scaling vs. Normalization

```python
# for Box-Cox Transformation
from scipy import stats

# for min_max scaling
from mlxtend.preprocessing import minmax_scaling
```

```python
copy_data2 = data.copy()
```

<span style="font-family: Arial; font-weight:bold;font-size:1.2em;color:#03b7b4;">1.12a) Check the Scaling data?

```python
# select the usd_goal_real column
usd_goal = copy_data2.Purchase
scaled_data = minmax_scaling(usd_goal, columns = [0])

# scale the goals from 0 to 1

# plot the original & scaled data together to compare
fig, ax=plt.subplots(1,2)
sns.distplot(copy_data2.Purchase, ax=ax[0])
ax[0].set_title("Original Data")
sns.distplot(scaled_data, ax=ax[1])
ax[1].set_title("Scaled data")
```

<span style="font-family: Arial; font-weight:bold;font-size:1.2em;color:#03b7b4;">1.12a) Let check Normalized data

```python
# get the index of all positive pledges (Box-Cox only takes postive values)
index_of_positive_pledges = copy_data2.Purchase > 0

# get only positive pledges (using their indexes)
positive_pledges = copy_data2.Purchase.loc[index_of_positive_pledges]

# normalize the pledges (w/ Box-Cox)
normalized_pledges = stats.boxcox(positive_pledges)[0]

# plot both together to compare
fig, ax=plt.subplots(1,2)
sns.distplot(positive_pledges, ax=ax[0])
ax[0].set_title("Original Data")
sns.distplot(normalized_pledges, ax=ax[1])
ax[1].set_title("Normalized data")

```

Read the raw data :https://raw.githubusercontent.com/selva86/datasets/master/a10.csv

```python
import statsmodels.api as sm
# Time series data source: 
import matplotlib.pyplot as plt
df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/a10.csv', parse_dates=['date'], index_col='date')

# Draw Plot
def plot_df(df, x, y, title="", xlabel='Date', ylabel='Value', dpi=100):
    plt.figure(figsize=(16,5), dpi=dpi)
    plt.plot(x, y, color='tab:red')
    plt.gca().set(title=title, xlabel=xlabel, ylabel=ylabel)
    plt.show()

plot_df(df, x=df.index, y=df.value, title='Monthly anti-diabetic drug sales in Australia from 1992 to 2008.')    
```
