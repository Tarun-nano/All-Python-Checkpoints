# merge function()

merge.inner join ==== returns the intersection of two dataframes

merge.outer join ==== returns the complete dataframe after combining both of them

merge.left join ==== returns the intersection as well as the left dataframe

merge.right join ==== returns the intersection as well as the right dataframe

syntax : pd.merge(left database,right database,on="key,how="inner")

```python
import pandas as pd
import numpy as np

h=["N","A","M","E"]
g=np.array(h)
h=pd.MultiIndex.from_product([["My DataFrame"],["3_months","6_months"],["python","ML"]])
z=([[12,34,57,43],[12,62,24,56],[54,21,89,69],[32,64,76,85]])
x=pd.DataFrame(z,index=[g,["Tarun","Rahul","Ravi","sid"]],columns=h)
print(x)


print()
print("stack()")
print(x.stack())
print()
print()
print("stack(level 0)")
print()
print(x.stack(0))
print()
print()
print("stack(level 1)")
print()
print(x.stack(1))
print()
print()
print("stack(level 2)")
print()
print(x.stack(2))
print()
print()
print("unstack()")
print(x.unstack())
print()
print()
print("unstack(level 1)")
print(x.unstack(1))
print()
print()
```
