```python
data=hr[["education","is_promoted"]]
```

```python
# Applying annova
import scipy.stats as stats

import statsmodels.api as sms
import statsmodels.formula.api as statsmodels
from statsmodels.formula.api import ols
import seaborn as sns
from statsmodels.stats.multicomp import pairwise_tukeyhsd
n = 3
dfn = n-1

dfd = data["education"].shape[0]
```

```python
stats.f.ppf(q=1-0.05, dfn=dfn, dfd=dfd)
```

```python
model_hr = ols("is_promoted~education",data=data).fit()
```

```python
print(sms.stats.anova_lm(model_hr))
```

```python
# the p value found out to be very low than 0.05, hence we can say that the education statistically doesnt have any relavively impact on the targer variable
```

```python
print(pairwise_tukeyhsd(data["is_promoted"],data["education"]))
```
