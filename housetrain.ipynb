# Check the linearity

h0: Data is linear

h1:data is not linear
    

```python
import statsmodels.api as sms
y=dummytrain.SalePrice
x=dummytrain.drop(columns="SalePrice")
X=dummytest
```

```python
model=sms.OLS(y,x).fit()
```

```python
model.summary()
```

```python
sms.stats.diagnostic.linear_rainbow(model)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y=train_test_split(x,y,random_state=1,test_size=.20)
lr.fit(pd.DataFrame(train_x),train_y)
```

```python
y_hat=lr.predict(pd.DataFrame(test_x))
y_hat
```

```python
# R square and RMSE

from sklearn.metrics import r2_score, mean_squared_error
print(np.sqrt(mean_squared_error(test_y,y_hat))) 
```

```python
print(r2_score(test_y,y_hat))                        #R Square
```
