# Q9. Groupby the data based on average ratings using bin edges

```python
newdata.describe()
```

```python
bin_edges=[0.00,3.77,3.96,4.13,5.30]
bin_names=["low","average","good","best"]
newdata["avg_ratings"]=pd.cut(newdata["average_rating"],bin_edges,labels=bin_names)
newdata.head()
```

```python
newdata2=newdata.groupby("avg_ratings")
newdata2.first()
```
