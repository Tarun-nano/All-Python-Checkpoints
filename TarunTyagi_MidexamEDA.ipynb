```python
data3=data[data["Subscribers"] != "--"]
```

```python
data4= data3[data3["Subscribers"] != " --"]
```

```python
 data5=data4[data4["Subscribers"] != "-- "]
```

```python
median=data5["Subscribers"].median()
median
```

```python
data["Subscribers"].replace(["--"," --","-- "],median,inplace=True)
```

```python
Sub_new=[]
for i in data["Subscribers"]:
    Sub_new.append(int(i))
data["Sub_new"]=Sub_new    
```

```python
data.sort_values("Sub_new",ascending=False).head(10)
```
